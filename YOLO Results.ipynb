```python
import cv2
from ultralytics import YOLO
import numpy as np

# Cargar el modelo YOLO preentrenado
model = YOLO("C:/Users/Johan/Desktop/YoloV8/best.pt")

# Realizar la predicci칩n en una imagen
results = model("C:/Users/Johan/Desktop/YoloV8/CarImages/prueba-multiple.jpg", conf=0.3, iou=0.5)

# Iterar sobre cada resultado en la lista
for result in results:
    # Obtener las coordenadas de los bounding boxes
    boxes = result.boxes.xyxy
    
    # Cargar la imagen
    image = cv2.imread("C:/Users/Johan/Desktop/YoloV8/CarImages/prueba-multiple.jpg")
    
    # Dibujar los bounding boxes en la imagen
    for box in boxes:
        xmin, ymin, xmax, ymax = box[:4] # Obtener las coordenadas del bounding box
        xmin, ymin, xmax, ymax = int(xmin), int(ymin), int(xmax), int(ymax)
        cv2.rectangle(image, (xmin, ymin), (xmax, ymax), (0,255,0), 2)
        cv2.putText(image, 'vehicle', (xmin, ymin - 5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

    # Mostrar la imagen
    cv2.imshow('Imagenes', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```
